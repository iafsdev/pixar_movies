# Exploratory Questions

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
df_movies = pd.read_csv('../data/processed/pixar_movies_complete.csv')
df_movies
```

## Function for Order Movies and Plot

```python
def order_movies(column):
  df = df_movies.sort_values(by=column, ascending=False)
  ax = sns.barplot(df, x='Movie', y=column, palette='flare')
  ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

## Movies Ordered by Worlwide Gross

```python
order_movies('Worldwide Gross')
```

## Movies Ordered by Rating

### Rotten Tomatoes

```python
order_movies('RT Score')
```

### IMDb

```python
order_movies('IMDB Score')
```

### Metacritic

```python
order_movies('Metacritic Score')
```

### Mean

```python
df_movies['IMDB Score'] = df_movies['IMDB Score'] * 10
df_movies
```

```python
df_movies['Score'] = (df_movies['RT Score'] + df_movies['IMDB Score'] + df_movies['Metacritic Score']) / 3
df_movies
```

```python
order_movies('Score')
```

## Movies Ordered by Domestic Gross

```python
order_movies('Domestic Gross')
```

## Movies Ordered by International Gross

```python
order_movies('International Gross')
```

## Correlations

```python
fig = plt.figure(figsize=(15,15))
sns.heatmap(df_movies.corr(), annot=True)
```

```python
order_movies('Oscars Nominated')
```
