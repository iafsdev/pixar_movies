## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
df_movies = pd.read_csv('../data/processed/pixar_movies_complete.csv')
df_movies
```

## Movies Ordered by Worlwide Gross

```python
df_movies_worldwide_gross = df_movies.sort_values(by='Worldwide Gross', ascending=False)
df_movies_worldwide_gross
```

```python
ax = sns.barplot(df_movies_worldwide_gross, x='Movie', y='Worldwide Gross', palette='flare')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```
