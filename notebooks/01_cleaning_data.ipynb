## Changing the Datatype of Domestic and International % from object to float

```python
df['Domestic %'] = df['Domestic %'].str.replace('%', '').astype(float)
df['International %'] = df['International %'].str.replace('%', '').astype(float)
df.info()
```

## Imputing the missing values

In this case, I insert the values because I know them

```python
df.loc[14, 'Oscars Nominated'] = 2
df.loc[14, 'Oscars Won'] = 1
df
```
