# Adding Movies

## Import libraries

```python
import pandas as pd
```

## Removing the Adjusted Domestic Gross column

```python
df_movies = pd.read_csv('../data/processed/PixarMovies.csv')
df_movies
```

```python
df_movies.drop('Adjusted Domestic Gross', axis=1, inplace=True)
df_movies.info()
```

## Adding the other movies

```python
df_other_movies = pd.read_csv('../data/raw/movies_pixar2.csv')
df_other_movies.info()
```

```python
df_movies = pd.concat([df_movies, df_other_movies], ignore_index=True)
df_movies
```

## Saving the new data

```python
df_movies.to_csv('../data/processed/pixar_movies_complete.csv', index=False)
```
